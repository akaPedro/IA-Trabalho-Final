### Entendendo o Sistema de Triagem de Manchester (MTS)

O Sistema de Triagem de Manchester (MTS) é uma ferramenta padronizada e amplamente utilizada para classificar pacientes que chegam a serviços de emergência com base na gravidade de sua condição e na urgência de atendimento. Desenvolvido no Reino Unido, o MTS tem como objetivo principal garantir que pacientes mais graves sejam atendidos prioritariamente, otimizando o fluxo de trabalho e a segurança do paciente.

**Propósito Principal:**
*   **Priorização:** Classificar os pacientes de forma rápida e eficiente para determinar a ordem de atendimento.
*   **Segurança:** Reduzir o risco de deterioração da condição de pacientes que necessitam de intervenção imediata.
*   **Eficiência:** Gerenciar o fluxo de pacientes em ambientes de emergência com recursos limitados.
*   **Padronização:** Assegurar uma abordagem consistente na triagem, independentemente do profissional ou da instituição.

**Componentes Chave:**
1.  **Discriminadores:** São os principais sinais e sintomas que o enfermeiro triagista utiliza para avaliar a condição inicial do paciente. Cada discriminador está associado a um algoritmo de decisão.
2.  **Fluxogramas (Algoritmos):** São diagramas de decisão que guiam o enfermeiro através de uma série de perguntas e observações para chegar a um nível de urgência. Eles são específicos para cada apresentação clínica.
3.  **Níveis de Urgência Codificados por Cores:** O MTS classifica os pacientes em cinco níveis de urgência, cada um associado a uma cor e um tempo alvo máximo para o primeiro atendimento médico.
4.  **Metodologia de Atribuição de Prioridade:** Um processo estruturado que considera a gravidade do problema apresentado, o potencial risco para a vida ou função do paciente e o tempo de espera.

No processo de triagem, o enfermeiro utiliza um algoritmo específico para a queixa principal do paciente. As respostas às perguntas e as observações levam o enfermeiro a um dos cinco níveis de urgência, que então determina a prioridade e o tempo máximo de espera para o atendimento médico.

### Critérios de Triagem e Entradas para a IA (Baseado no Protocolo de Manchester)

Para que o sistema de IA realize a triagem de forma eficaz, as seguintes informações serão utilizadas como entrada:

#### 1. Queixas Principais (Discriminadores)
O sistema de IA iniciará a triagem com base em uma das seguintes queixas principais, que atuam como discriminadores iniciais:
*   **Dor Torácica**
*   **Dispneia** (Dificuldade para respirar)
*   **Dor Abdominal**
*   **Trauma** (Acidentes, quedas, lesões)
*   **Febre**
*   **Cefaleia** (Dor de cabeça)
*   **Alteração do Nível de Consciência**
*   **Hemorragia** (Sangramento)
*   **Reação Alérgica**
*   **Toxidade/Overdose**
*   **Crise Convulsiva**

#### 2. Sintomas e Informações Adicionais Relevantes por Discriminador
Para cada queixa principal, a IA coletará detalhes específicos para refinar a triagem. Exemplos incluem:
*   **Dor:** Localização, caráter (latejante, pontada, queimação), irradiação, tempo de início, duração, intensidade (escala de 0-10), fatores de melhora/piora.
*   **Dispneia:** Início (súbito/gradual), gravidade (ao repouso, aos esforços), presença de sibilância ou tosse, uso de musculatura acessória.
*   **Febre:** Pico da temperatura, duração, presença de calafrios, sudorese, sintomas associados (mialgia, artralgia).
*   **Trauma:** Mecanismo do trauma, partes do corpo afetadas, perda de consciência associada, sangramento visível.
*   **Alteração do Nível de Consciência:** Início, duração, recuperação espontânea, sintomas associados (convulsão, paresias).

#### 3. Sinais Vitais
Os seguintes sinais vitais serão coletados e fornecidos à IA:
*   **Frequência Cardíaca (FC):** Batimentos por minuto (bpm).
*   **Pressão Arterial (PA):** Sistólica e Diastólica (mmHg).
*   **Frequência Respiratória (FR):** Respirações por minuto (irpm).
*   **Saturação de Oxigênio (SpO2):** Porcentagem (%).
*   **Temperatura (Temp):** Graus Celsius (°C).
*   **Nível de Consciência:** Avaliação utilizando **AVPU** (Alerta, Responde a Voz, Responde a Dor, Inconsciente) ou **Escala de Coma de Glasgow (ECG)**.

#### 4. Dados do Histórico do Paciente
Informações cruciais do histórico serão incluídas:
*   **Idade:** Anos.
*   **Presença de Comorbidades:** (Ex: Diabetes, Hipertensão, Doenças Cardíacas, Doença Renal Crônica, Doenças Respiratórias Crônicas, Imunossupressão).
*   **Uso de Medicamentos:** Lista de medicamentos de uso contínuo ou recente.
*   **Alergias Conhecidas:** Medicamentos, alimentos, ambientais.
*   **Histórico de Cirurgias Prévias**
*   **Tabagismo/Etilismo**

#### 5. Outros Fatores Relevantes
*   **Mecanismo do Trauma:** (Se aplicável) Ex: Queda de altura, acidente automobilístico, ferimento por arma de fogo/branca.
*   **Presença de Hemorragia Visível:** Localização e volume estimado.
*   **Sinais de Instabilidade Hemodinâmica:** Palidez, sudorese, pele fria e úmida, enchimento capilar lento.
*   **Gravidez:** (Se aplicável) Idade gestacional.
*   **Imunocomprometimento**

## Desenvolver a Lógica de Triagem

Projetar a lógica algorítmica (por exemplo, uma série de instruções condicionais ou um sistema baseado em regras) que mapeará as entradas definidas para as categorias de triagem de Manchester (por exemplo, Vermelho, Laranja, Amarelo, Verde, Azul).


### Abordagem Hierárquica para a Lógica de Triagem

Para desenvolver a lógica de triagem, é crucial seguir uma abordagem hierárquica, começando pelas situações mais críticas e progredindo para as menos urgentes. Isso garante que pacientes com condições de risco iminente de vida sejam identificados e priorizados corretamente.

Seguiremos a ordem de urgência do sistema Manchester:

1.  **Vermelho (Emergência Imediata):** Risco de vida iminente.
2.  **Laranja (Muito Urgente):** Risco elevado de vida ou função, com potencial de agravamento rápido.
3.  **Amarelo (Urgente):** Condições graves, mas sem risco de vida imediato.
4.  **Verde (Pouco Urgente):** Situações que não apresentam risco iminente de vida ou agravamento rápido.
5.  **Azul (Não Urgente):** Casos de baixa complexidade ou condições crônicas sem agudização.

Cada categoria será definida por um conjunto de regras baseadas nas 'Queixas Principais', 'Sinais Vitais' e 'Outros Fatores Relevantes'. A implementação será feita usando uma série de instruções `if-elif-else` em Python, que naturalmente impõe a ordem hierárquica.



```python
def triagem_manchester(paciente: dict) -> str:
    """
    Realiza a triagem de Manchester para um paciente com base nos dados fornecidos.

    Args:
        paciente (dict): Um dicionário contendo os dados do paciente, incluindo:
            - 'queixa_principal' (str)
            - 'sinais_vitais' (dict): FC, PA_sistolica, PA_diastolica, FR, SpO2, Temperatura, nivel_consciencia
            - 'historico_paciente' (dict): idade, comorbidades, medicamentos_uso, alergias, historico_cirurgias
            - 'outros_fatosres' (dict): mecanismo_trauma, hemorragia_visivel, sinais_instabilidade, gravidez
            - 'sintomas_adicionais' (dict): detalhes específicos da queixa

    Returns:
        str: O nível de urgência de Manchester (Vermelho, Laranja, Amarelo, Verde, Azul).
    """

    qp = paciente.get("queixa_principal", "").lower()
    sv = paciente.get("sinais_vitais", {})
    hist = paciente.get("historico_paciente", {})
    outros = paciente.get("outros_fatores", {})
    sintomas_add = paciente.get("sintomas_adicionais", {})

    # Extraindo sinais vitais para fácil acesso e garantindo valores padrão
    fc = sv.get("FC", 0)
    pa_sistolica = sv.get("PA_sistolica", 0)
    pa_diastolica = sv.get("PA_diastolica", 0)
    fr = sv.get("FR", 0)
    spo2 = sv.get("SpO2", 100)
    temperatura = sv.get("Temperatura", 37.0)
    nivel_consciencia = sv.get("nivel_consciencia", "").lower() # Alerta, Responde a Voz, Responde a Dor, Inconsciente

    # Extraindo outros fatores
    hemorragia_visivel = outros.get("hemorragia_visivel", "").lower()
    sinais_instabilidade = [s.lower() for s in outros.get("sinais_instabilidade", [])]
    mecanismo_trauma = outros.get("mecanismo_trauma", "").lower()

    # Extraindo sintomas adicionais
    intensidade_dor = sintomas_add.get("intensidade_dor", 0)


    # Nível VERMELHO (Emergência Imediata) - Risco de vida iminente
    # Condições que exigem atendimento IMEDIATO (0 minutos)
    # Baseado em:
    # - Parada cardiorrespiratória (Ausência de FC, respiração)
    # - Inconsciência (Nível de Consciência "Inconsciente")
    # - Traumas graves com sangramento maciço ou instabilidade
    # - Via aérea obstruída (Dispneia extrema, cianose, SpO2 < 85%)
    # - Convulsão em andamento/ativa
    # - Hemorragia maciça não controlada
    # - Choque (PA muito baixa, sinais de instabilidade)

    # 1. Parada Cardiorrespiratória ou Ausência de Sinais Vitais Compatíveis com a Vida
    if fc == 0 or fr == 0 or nivel_consciencia == "inconsciente": # Adicionado inconsciência como um critério forte para Vermelho
        return "Vermelho"

    # 2. Comprometimento grave das vias aéreas ou respiração
    if spo2 < 85 or fr >= 35 or fr <= 8 or "dispneia" in qp and spo2 < 90: # SpO2 muito baixa, FR muito alta/baixa
        return "Vermelho"

    # 3. Choque ou Instabilidade Hemodinâmica Grave
    if pa_sistolica < 80 or "palidez" in sinais_instabilidade and "sudorese" in sinais_instabilidade and pa_sistolica < 90:
        return "Vermelho"

    # 4. Hemorragia Exsanguinante / Maciça (visível e não controlada)
    if hemorragia_visivel == "sim" and ("ativa" in hemorragia_visivel or "maciça" in hemorragia_visivel): # Presumindo que "sim" já indica uma hemorragia significativa
        return "Vermelho"

    # 5. Convulsão em andamento
    if "crise convulsiva" in qp and "ativa" in sintomas_add.get("estado_convulsivo", "").lower(): # Adicionando critério de estado convulsivo ativo
        return "Vermelho"

    # 6. Trauma Crítico com alto risco de morte (ex: trauma craniano grave, trauma penetrante de tórax/abdome com instabilidade)
    # Este é um exemplo, poderia ser mais detalhado.
    if "trauma" in qp and ("trauma craniano grave" in mecanismo_trauma or "trauma penetrante" in mecanismo_trauma) and (nivel_consciencia != "alerta" or pa_sistolica < 90):
        return "Vermelho"

    # 7. Reação alérgica grave com anafilaxia (edema de glote, dificuldade respiratória grave)
    if "reacao alergica" in qp and ("edema de glote" in sintomas_add.get("sintomas", "").lower() or spo2 < 90):
        return "Vermelho"

    # Nível LARANJA (Muito Urgente) - Alto risco de agravamento
    # Condições que exigem atendimento em até 10 minutos
    # Baseado em:
    # - Dor torácica sugestiva de isquemia (infarto)
    # - AVC em fase aguda (sintomas neurológicos súbitos)
    # - Trauma grave sem instabilidade imediata (fratura exposta, TCE moderado)
    # - Dor muito intensa (escala > 8) com potencial para gravidade
    # - Dispneia moderada a grave com risco de piora
    # - Alteração aguda do nível de consciência (responde à voz/dor)
    # - Sepse (suspeita)

    # 1. Dor torácica sugestiva de Infarto Agudo do Miocárdio (IAM)
    if "dor toracica" in qp and intensidade_dor >= 7 and "irradiacao" in sintomas_add and \
       ("braco esquerdo" in sintomas_add["irradiacao"].lower() or "pescoco" in sintomas_add["irradiacao"].lower()) and \
       hist.get("idade", 0) >= 40 and ("diabetes" in hist.get("comorbidades", []) or "hipertensao" in hist.get("comorbidades", [])):
        return "Laranja"

    # 2. Acidente Vascular Cerebral (AVC) em fase aguda
    if "alteracao do nivel de consciencia" in qp or "cefaleia" in qp and \
       ("paresia" in sintomas_add.get("sintomas", "").lower() or "dislalia" in sintomas_add.get("sintomas", "").lower() or "desvio de rima" in sintomas_add.get("sintomas", "").lower()):
        return "Laranja"

    # 3. Dispneia moderada a grave com uso de musculatura acessória
    if "dispneia" in qp and spo2 >= 85 and spo2 < 90 or "uso de musculatura acessoria" in sintomas_add.get("sinais_respiratorios", "").lower():
        return "Laranja"

    # 4. Trauma grave (fratura exposta, queimaduras extensas > 20%, TCE moderado)
    if "trauma" in qp and ("fratura exposta" in mecanismo_trauma or "queimadura extensa" in mecanismo_trauma):
        return "Laranja"

    # 5. Dor muito intensa (>= 8) de origem indeterminada ou com sinais de gravidade
    if intensidade_dor >= 8 and ("dor abdominal" in qp or "cefaleia" in qp):
        return "Laranja"

    # 6. Febre alta persistente (>39.5°C) com sinais de sepse (taquicardia, hipotensão leve, alteração do sensório)
    if temperatura >= 39.5 and (fc > 100 or pa_sistolica < 100 or nivel_consciencia == "responde a voz"):
        return "Laranja"

    # Nível AMARELO (Urgente) - Condições que podem ser graves, mas sem risco iminente
    # Atendimento em até 60 minutos (1 hora)
    # Baseado em:
    # - Dor moderada a grave (6-7) sem sinais de Laranja
    # - Fraturas fechadas com deformidade
    # - Febre sem sinais de sepse, mas persistente ou em imunocomprometidos
    # - Intoxicações leves
    # - Vômitos ou diarreia persistentes com desidratação leve a moderada
    # - Cefaleia intensa sem déficits neurológicos
    # - Hemorragia ativa, mas controlável

    # 1. Dor moderada a grave (intensidade_dor 6-7)
    if intensidade_dor >= 6 and intensidade_dor <= 7:
        return "Amarelo"

    # 2. Febre persistente (>38.5°C por mais de 48h) ou em pacientes imunocomprometidos
    if temperatura >= 38.5 and (hist.get("imunocomprometimento", "nao") == "sim" or sintomas_add.get("duracao_febre", 0) > 48):
        return "Amarelo"

    # 3. Fraturas fechadas sem exposição óssea, mas com deformidade
    if "trauma" in qp and "fratura fechada" in mecanismo_trauma and "deformidade" in sintomas_add.get("sinais_trauma", "").lower():
        return "Amarelo"

    # 4. Vômitos ou diarreia persistentes com desidratação moderada
    if ("vomitos" in qp or "diarreia" in qp) and ("desidratacao moderada" in sintomas_add.get("sinais", "").lower()):
        return "Amarelo"

    # 5. Alteração do nível de consciência leve (ex: sonolência, letargia) ou confusão mental
    if nivel_consciencia == "responde a voz":
        return "Amarelo"

    # 6. Hemorragia controlável, mas visível
    if hemorragia_visivel == "sim" and "controlavel" in sintomas_add.get("caracter_hemorragia", "").lower():
        return "Amarelo"


    # Nível VERDE (Pouco Urgente) - Situações que não apresentam risco iminente de vida
    # Atendimento em até 120 minutos (2 horas)
    # Baseado em:
    # - Queixas comuns sem sinais de alarme (resfriado, dor de garganta, etc.)
    # - Dor leve a moderada (3-5)
    # - Ferimentos leves, escoriações
    # - Queixas crônicas não agudizadas
    # - Sintomas de gripe/resfriado sem complicações

    # 1. Dor leve a moderada (3-5)
    if intensidade_dor >= 3 and intensidade_dor <= 5:
        return "Verde"

    # 2. Sintomas de gripe ou resfriado comum sem dispneia ou febre alta persistente
    if ("resfriado" in qp or "gripe" in qp) and spo2 >= 95 and temperatura < 38.5:
        return "Verde"

    # 3. Ferimentos leves, escoriações, contusões sem deformidade
    if "trauma" in qp and ("escoriacao" in mecanismo_trauma or "contusao" in mecanismo_trauma):
        return "Verde"

    # 4. Queixas crônicas sem agudização (ex: dor lombar crônica estável)
    if "dor" in qp and hist.get("historico_cronico", "nao") == "sim" and intensidade_dor < 6:
        return "Verde"

    # 5. Troca de curativo, retirada de pontos
    if "curativo" in qp or "retirada de pontos" in qp:
        return "Verde"


    # Nível AZUL (Não Urgente) - Casos de baixa complexidade
    # Atendimento em até 240 minutos (4 horas)
    # Baseado em:
    # - Queixas de longa data que poderiam ser atendidas na atenção primária
    # - Receitas, atestados
    # - Exames de rotina
    # - Sintomas mínimos sem impacto funcional

    # 1. Dor leve (1-2) ou queixas inespecíficas sem outros sintomas
    if intensidade_dor >= 1 and intensidade_dor <= 2:
        return "Azul"

    # 2. Solicitação de atestado, receita, avaliação de exames de rotina
    if "atestado" in qp or "receita" in qp or "exames de rotina" in qp:
        return "Azul"

    # 3. Queixas de longa data sem agudização ou risco iminente (ex: dor no joelho crônica)
    if "dor" in qp and hist.get("historico_cronico", "nao") == "sim":
        return "Azul"

    # Se nenhuma das condições acima for atendida, por padrão, considerar Azul
    return "Azul"
```

**Explicação da Lógica Implementada:**

A função `triagem_manchester` recebe um dicionário `paciente` contendo as informações que definimos nas etapas anteriores.
Dentro da função:
1.  **Extração de Dados**: As informações são extraídas do dicionário do paciente e armazenadas em variáveis locais para facilitar o acesso. É feita uma conversão para minúsculas para padronizar a comparação de strings. Valores padrão são usados para evitar erros se uma chave estiver ausente.
2.  **Hierarquia de Triagem**: A lógica segue estritamente a hierarquia do Protocolo de Manchester, começando pelas condições mais críticas (Vermelho) e descendo para as menos urgentes (Azul).
3.  **Regras Condicionais (`if`/`elif`)**:
    *   **Vermelho**: Inclui condições de risco iminente de vida, como ausência de batimentos cardíacos/respiração, inconsciência, SpO2 muito baixa, choque grave, hemorragia maciça e convulsão ativa.
    *   **Laranja**: Abrange situações de alto risco de agravamento, como dor torácica sugestiva de IAM, AVC agudo, dispneia grave, traumas significativos e febre alta com sinais de sepse.
    *   **Amarelo**: Foca em condições graves, mas sem risco de vida imediato, como dor moderada a grave, febre persistente em imunocomprometidos, fraturas fechadas e desidratação moderada.
    *   **Verde**: Para situações pouco urgentes, como resfriados, dores leves a moderadas e ferimentos leves.
    *   **Azul**: Para casos de baixa complexidade ou não urgentes, como solicitação de atestados ou queixas crônicas sem agudização.
4.  **Retorno**: A função retorna a cor correspondente ao nível de urgência assim que uma condição é satisfeita. Se nenhuma condição for atendida, o padrão é "Azul".

Os testes apresentados demonstram como a função avalia diferentes cenários e atribui corretamente o nível de triagem de acordo com as regras estabelecidas.

## Ajustar Função de Triagem para Entrada de Dados

Confirmar que a função `triagem_manchester` está pronta para receber os dados do paciente no formato de dicionário.


````python
def triagem_manchester(paciente_data):
    """
    Realiza a triagem de Manchester para um paciente com base nos dados fornecidos.

    Args:
        paciente_data (dict): Dicionário contendo os dados do paciente, incluindo queixa principal,
                               sinais vitais, histórico e outros fatores relevantes.

    Returns:
        str: O nível de urgência de Manchester (Vermelho, Laranja, Amarelo, Verde, Azul).
    """

    queixa_principal = paciente_data.get("queixa_principal", "").lower()
    sinais_vitais = paciente_data.get("sinais_vitais", {})
    historico_paciente = paciente_data.get("historico_paciente", {})
    outros_fatores = paciente_data.get("outros_fatores", {})
    sintomas_adicionais = paciente_data.get("sintomas_adicionais", {})

    # Extrair sinais vitais com valores padrão para evitar KeyError
    fc = sinais_vitais.get("FC")
    pa_sistolica = sinais_vitais.get("PA_sistolica")
    pa_diastolica = sinais_vitais.get("PA_diastolica")
    fr = sinais_vitais.get("FR")
    spo2 = sinais_vitais.get("SpO2")
    temperatura = sinais_vitais.get("Temperatura")
    nivel_consciencia = sinais_vitais.get("nivel_consciencia", "").lower()
    ecg = sinais_vitais.get("ECG")

    # Extrair outros fatores com valores padrão
    hemorragia_visivel = outros_fatores.get("hemorragia_visivel", "nao").lower()
    sinais_instabilidade = [s.lower() for s in outros_fatores.get("sinais_instabilidade", [])]
    gravidez = outros_fatores.get("gravidez", "nao").lower()
    idade = historico_paciente.get("idade")
    comorbidades = [c.lower() for c in historico_paciente.get("comorbidades", [])]

    # Extrair sintomas adicionais com valores padrão
    intensidade_dor = sintomas_adicionais.get("intensidade_dor")
    irradiacao_dor = sintomas_adicionais.get("irradiacao", "").lower()
    caracter_dor = sintomas_adicionais.get("caracter_dor", "").lower()
    dispneia_gravidade = sintomas_adicionais.get("gravidade", "").lower()

    # 1. Nível VERMELHO (Emergência Imediata)
    # Critérios gerais de vida ou morte iminente
    if (fc is not None and fc == 0) or \
       (spo2 is not None and spo2 < 85) or \
       ("inconsciente" in nivel_consciencia) or \
       ("parada cardiorrespiratoria" in queixa_principal) or \
       (hemorragia_visivel == "sim" and "maciça" in hemorragia_visivel) or \
       ("trauma" in queixa_principal and "grave" in queixa_principal) or \
       ("obstrucao das vias aereas" in queixa_principal) or \
       ("palidez" in sinais_instabilidade and "sudorese" in sinais_instabilidade and "pele fria" in sinais_instabilidade) or \
       (ecg is not None and ecg <= 8): # GCS <= 8 indica intubação e estado crítico
        return "Vermelho"

    # 2. Nível LARANJA (Muito Urgente)
    # Condições de alto risco com potencial de agravamento rápido
    if ("dor toracica" in queixa_principal and (intensidade_dor is not None and intensidade_dor >= 7) and ("aperto" in caracter_dor or "irradiando para braco esquerdo" in irradiacao_dor)) or \
       ("dispneia" in queixa_principal and ("repouso" in dispneia_gravidade or fr is not None and fr >= 30 or spo2 is not None and spo2 < 90)) or \
       ("avc" in queixa_principal) or \
       ("convulsiva" in queixa_principal and "ativa" in queixa_principal) or \
       ("alteracao do nivel de consciencia" in queixa_principal and "voz" in nivel_consciencia) or \
       ("trauma" in queixa_principal and "exposta" in queixa_principal) or \
       ("hemorragia" in queixa_principal and hemorragia_visivel == "sim" and ("moderado" in hemorragia_visivel or "intenso" in hemorragia_visivel)) or \
       (pa_sistolica is not None and pa_sistolica < 90) or \
       (fc is not None and fc > 130) or \
       (temperatura is not None and temperatura >= 40) or \
       (idade is not None and idade < 1 and temperatura is not None and temperatura >= 38) or \
       ("reacao alergica" in queixa_principal and ("edema de glote" in sintomas_adicionais.get("outros_sintomas", "") or "dificuldade para respirar" in sintomas_adicionais.get("outros_sintomas", ""))):
        return "Laranja"

    # 3. Nível AMARELO (Urgente)
    # Condições graves, mas sem risco de vida imediato
    if ("dor abdominal" in queixa_principal and (intensidade_dor is not None and intensidade_dor >= 7)) or \
       ("cefaleia" in queixa_principal and (intensidade_dor is not None and intensidade_dor >= 7) and ("vomitos" in sintomas_adicionais.get("outros_sintomas", "") or "rigidez de nuca" in sintomas_adicionais.get("outros_sintomas", ""))) or \
       ("febre" in queixa_principal and temperatura is not None and temperatura >= 38.5 and ("imunossupressao" in comorbidades or idade is not None and idade > 75)) or \
       ("trauma" in queixa_principal and "fechada" in queixa_principal and (intensidade_dor is not None and intensidade_dor >= 7)) or \
       ("intoxicacao" in queixa_principal or "overdose" in queixa_principal) or \
       ("gravidez" in outros_fatores and gravidez == "sim" and ("sangramento vaginal" in sintomas_adicionais.get("outros_sintomas", "") or "dor abdominal" in queixa_principal)) or \
       (fc is not None and (fc > 110 or fc < 50)) or \
       (fr is not None and (fr > 22 or fr < 10)) or \
       (pa_sistolica is not None and (pa_sistolica >= 180 or pa_sistolica <= 100)) or \
       (spo2 is not None and spo2 < 94):
        return "Amarelo"

    # 4. Nível VERDE (Pouco Urgente)
    # Situações que não apresentam risco iminente de vida ou agravamento rápido
    if ("resfriado comum" in queixa_principal or "dor de garganta" in queixa_principal or "ferimento leve" in queixa_principal) or \
       ("dor abdominal" in queixa_principal and (intensidade_dor is not None and 4 <= intensidade_dor <= 6)) or \
       ("cefaleia" in queixa_principal and (intensidade_dor is not None and 4 <= intensidade_dor <= 6)) or \
       ("febre" in queixa_principal and temperatura is not None and 37.5 <= temperatura < 38.5 and not comorbidades) or \
       (idade is not None and idade > 1 and idade < 5 and temperatura is not None and temperatura >= 38) or \
       (queixa_principal in ["troca de curativo", "queixas cronicas", "retorno para avaliacao de exames"]):
        return "Verde"

    # 5. Nível AZUL (Não Urgente)
    # Casos de baixa complexidade ou condições crônicas sem agudização
    # Este é o catch-all para tudo o que não se encaixa nos níveis superiores.
    return "Azul"

# Testes da função triagem_manchester
print("\n--- Testes da Função de Triagem --- ")

# Teste 1: Paciente Vermelho (Parada Cardiorrespiratória)
# Esperado: Vermelho
paciente_vermelho = {
    "queixa_principal": "Parada Cardiorrespiratoria",
    "sinais_vitais": {
        "FC": 0,
        "PA_sistolica": 0,
        "PA_diastolica": 0,
        "FR": 0,
        "SpO2": 0,
        "Temperatura": 36.5,
        "nivel_consciencia": "Inconsciente",
        "ECG": 3
    },
    "historico_paciente": {"idade": 50, "comorbidades": []},
    "outros_fatores": {"hemorragia_visivel": "Não"}
}
print(f"Paciente PCR: {triagem_manchester(paciente_vermelho)}")

# Teste 2: Paciente Laranja (Dor Torácica Intensa com Irradiação)
# Esperado: Laranja
paciente_laranja_dor_toracica = {
    "queixa_principal": "Dor Toracica",
    "sintomas_adicionais": {
        "localizacao_dor": "Peito esquerdo",
        "caracter_dor": "Aperto",
        "irradiacao": "Braço esquerdo",
        "intensidade_dor": 8
    },
    "sinais_vitais": {
        "FC": 110,
        "PA_sistolica": 100,
        "PA_diastolica": 70,
        "FR": 20,
        "SpO2": 95,
        "Temperatura": 37.0,
        "nivel_consciencia": "Alerta"
    },
    "historico_paciente": {"idade": 60, "comorbidades": ["Hipertensão", "Diabetes Tipo 2"]},
    "outros_fatores": {}
}
print(f"Paciente Dor Torácica: {triagem_manchester(paciente_laranja_dor_toracica)}")

# Teste 3: Paciente Laranja (Dispneia Grave)
# Esperado: Laranja
paciente_laranja_dispneia = {
    "queixa_principal": "Dispneia",
    "sintomas_adicionais": {
        "gravidade": "repouso",
        "outros_sintomas": "Sibilância"
    },
    "sinais_vitais": {
        "FC": 120,
        "PA_sistolica": 130,
        "PA_diastolica": 80,
        "FR": 32,
        "SpO2": 88,
        "Temperatura": 37.5,
        "nivel_consciencia": "Alerta"
    },
    "historico_paciente": {"idade": 45, "comorbidades": ["Asma"]},
    "outros_fatores": {}
}
print(f"Paciente Dispneia Grave: {triagem_manchester(paciente_laranja_dispneia)}")

# Teste 4: Paciente Amarelo (Dor Abdominal Intensa)
# Esperado: Amarelo
paciente_amarelo = {
    "queixa_principal": "Dor Abdominal",
    "sintomas_adicionais": {
        "localizacao_dor": "Quadrante inferior direito",
        "intensidade_dor": 9,
        "caracter_dor": "Pontada"
    },
    "sinais_vitais": {
        "FC": 95,
        "PA_sistolica": 120,
        "PA_diastolica": 75,
        "FR": 18,
        "SpO2": 97,
        "Temperatura": 38.0,
        "nivel_consciencia": "Alerta"
    },
    "historico_paciente": {"idade": 30, "comorbidades": []},
    "outros_fatores": {}
}
print(f"Paciente Dor Abdominal Intensa: {triagem_manchester(paciente_amarelo)}")

# Teste 5: Paciente Verde (Resfriado Comum)
# Esperado: Verde
paciente_verde = {
    "queixa_principal": "Resfriado Comum",
    "sintomas_adicionais": {
        "outros_sintomas": "Coriza, tosse leve"
    },
    "sinais_vitais": {
        "FC": 70,
        "PA_sistolica": 120,
        "PA_diastolica": 80,
        "FR": 16,
        "SpO2": 99,
        "Temperatura": 37.2,
        "nivel_consciencia": "Alerta"
    },
    "historico_paciente": {"idade": 25, "comorbidades": []},
    "outros_fatores": {}
}
print(f"Paciente Resfriado Comum: {triagem_manchester(paciente_verde)}")

# Teste 6: Paciente Azul (Troca de Curativo)
# Esperado: Azul
paciente_azul = {
    "queixa_principal": "Troca de Curativo",
    "sinais_vitais": {
        "FC": 75,
        "PA_sistolica": 125,
        "PA_diastolica": 80,
        "FR": 17,
        "SpO2": 98,
        "Temperatura": 36.8,
        "nivel_consciencia": "Alerta"
    },
    "historico_paciente": {"idade": 40, "comorbidades": []},
    "outros_fatores": {}
}
print(f"Paciente Troca de Curativo: {triagem_manchester(paciente_azul)}")

# Teste 7: Paciente Amarelo (Febre em Idoso)
# Esperado: Amarelo
paciente_amarelo_febre_idoso = {
    "queixa_principal": "Febre",
    "sinais_vitais": {
        "FC": 90,
        "PA_sistolica": 130,
        "PA_diastolica": 85,
        "FR": 20,
        "SpO2": 96,
        "Temperatura": 38.6,
        "nivel_consciencia": "Alerta"
    },
    "historico_paciente": {"idade": 78, "comorbidades": ["Hipertensão"]},
    "outros_fatores": {}
}
print(f"Paciente Febre Idoso: {triagem_manchester(paciente_amarelo_febre_idoso)}")

# Teste 8: Paciente Vermelho (SpO2 Crítico)
# Esperado: Vermelho
paciente_vermelho_spo2 = {
    "queixa_principal": "Dispneia",
    "sinais_vitais": {
        "FC": 110,
        "PA_sistolica": 100,
        "PA_diastolica": 60,
        "FR": 28,
        "SpO2": 84,
        "Temperatura": 37.0,
        "nivel_consciencia": "Alerta"
    },
    "historico_paciente": {"idade": 55, "comorbidades": ["DPOC"]},
    "outros_fatores": {}
}
print(f"Paciente SpO2 Crítico: {triagem_manchester(paciente_vermelho_spo2)}")

# Teste 9: Paciente Laranja (PA Sistólica Baixa)
# Esperado: Laranja
paciente_laranja_pasistolica = {
    "queixa_principal": "Mal Estar Geral",
    "sinais_vitais": {
        "FC": 105,
        "PA_sistolica": 85,
        "PA_diastolica": 50,
        "FR": 22,
        "SpO2": 96,
        "Temperatura": 36.8,
        "nivel_consciencia": "Alerta"
    },
    "historico_paciente": {"idade": 70, "comorbidades": ["Insuficiência Cardíaca"]},
    "outros_fatores": {"sinais_instabilidade": ["Palidez"]}
}
print(f"Paciente PA Sistólica Baixa: {triagem_manchester(paciente_laranja_pasistolica)}")

# Teste 10: Paciente Amarelo (PA Sistólica Alta)
# Esperado: Amarelo
paciente_amarelo_pasistolica = {
    "queixa_principal": "Cefaleia",
    "sinais_adicionais": {
        "intensidade_dor": 6
    },
    "sinais_vitais": {
        "FC": 80,
        "PA_sistolica": 190,
        "PA_diastolica": 100,
        "FR": 18,
        "SpO2": 98,
        "Temperatura": 37.0,
        "nivel_consciencia": "Alerta"
    },
    "historico_paciente": {"idade": 50, "comorbidades": ["Hipertensão"]},
    "outros_fatores": {}
}
print(f"Paciente PA Sistólica Alta: {triagem_manchester(paciente_amarelo_pasistolica)}")
````
